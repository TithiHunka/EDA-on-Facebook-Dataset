# **6. Data Pre-Processing**
---



```
data.shape
```

```
#information of Missing data: 
null_frame = pd.DataFrame(index = data.columns.values)
null_frame['Null Frequency'] = data.isnull().sum().values
percent = data.isnull().sum().values/data.shape[0]
null_frame['Missing %age'] = np.round(percent, decimals = 4) * 100
null_frame.transpose()

#Observations:
#Approx 18% (175)missing values in gender so to replace it with mode
#2 missing values in Tenure we will replace it with mode
data['gender'] = data['gender'].replace(np.nan, data['gender'].mode()[0])
data['tenure'] = data['tenure'].replace(np.nan, data['tenure'].mode()[0])

#display with updations:
null_frame = pd.DataFrame(index = data.columns.values)
null_frame['Null Frequency'] = data.isnull().sum().values
percent = data.isnull().sum().values/data.shape[0]
null_frame['Missing %age'] = np.round(percent, decimals = 4) * 100
null_frame.transpose()
```

```
#skewness of 3 columns(likes_received, mobile_likes_received,www_likes_received)
data.hist(grid=False,  figsize=(10, 6),bins=30) #plot skewness
```

```
data.skew(axis = 0, skipna = True) #skewness value
```

```
data['likes_received_log']=np.log(data['likes_received']+1)
data['mobile_likes_received_log']=np.log(data['mobile_likes_received']+1)
data['www_likes_received_log']=np.log(data['www_likes_received']+1)
data['www_likes_log']=np.log(data['www_likes']+1)
data['mobile_likes_log']=np.log(data['mobile_likes']+1) #Skewnwss removed by log(x+1) Transformation
```

```
data.skew(axis = 0, skipna = True) #skewness value after skewness removal
```

# **8. Exploratory Data Analysis**
---


```
labels=['10-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100','101-110','111-120']
data['age_group'] = pd.cut(data.age,bins=np.arange(10,121,10),labels=labels,right=True)
data.sample()
```

```
import datetime
data['dateInt']=data['dob_year'].astype(str) + data['dob_month'].astype(str).str.zfill(2)+ data['dob_day'].astype(str).str.zfill(2)
data['Date'] = pd.to_datetime(data['dateInt'], format='%Y%m%d')
data
```

```
tenlabel=['0-1 year','1-2 years','2-3 years','3-4 years','4-5 years','5-6 years','6-7 years','7-8 years','8-9 years']
data['year_group']=pd.cut(data.tenure,bins=np.arange(0,3300,365),labels=tenlabel,right=True)
data
```

# **1) Analysis based on Tenure**

```
#Tenure 
#1)Number of users at different tenure groups
data.year_group.fillna(value='0-1 year',inplace=True)
data.year_group.value_counts(dropna=False)
```

```
#2)plotting number of users at different tenure groups
CountStatus = pd.value_counts(data['year_group'].values, sort=True)
print(CountStatus)
plt.ylabel("Tenure(Years)")
plt.xlabel("Number of Users")
plt.title("No. of Users in differet Tenur Groups  ")
CountStatus.plot.barh()
```

```
#3)Average tenure days for different age groups
print(type(data.groupby('age_group').mean()))
data.groupby('age_group').mean()
```

```
#4)plotting of number of days spent in facebook by different age groups
figure = plt.figure(figsize=[15, 7])
data.groupby(by='age_group')['tenure'].sum().plot.line(color='green',marker='o')
plt.xlabel(xlabel='Age_group', size=14)
plt.ylabel(ylabel='Tenure(days)', size=14)
plt.title(label='Tenure(days) by different age groups', size=12)
```

# **observations:**
1)Maximum users tenure is less than 1 year.     

2)Most of the new users joined in 1 year period.

3)Average tenure(days) of all the age groups is different and heighest for age group 20-30.

# **2) Analysis based on Users**

```
#users
#1)Number of users (Male and female ratio)
data['gender'].value_counts()
Countgender = pd.value_counts(data['gender'].values, sort=True)
print(Countgender)
mylabels=['Males','Females']
colors=[ 'lightskyblue','lightcoral']
Countgender.plot.pie(labels=mylabels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Male Female Ratio")
```

```
#2)Averge friends_count according to gender
figure = plt.figure(figsize=[10, 5])
data.groupby(by='gender')['friend_count'].sum().plot.bar(color='yellow')
print(data.groupby(by='gender')['friend_count'].sum())
plt.ylabel(ylabel='Friends_count', size=14)
plt.title(label='Average friends count of males and females', size=16)
```

```
#3)Averge friendships_initiated according to gender
figure = plt.figure(figsize=[10, 5])
data.groupby(by='gender')['friendships_initiated'].sum().plot.bar(color='blue')
print(data.groupby(by='gender')['friendships_initiated'].sum())
plt.ylabel(ylabel='Friendship_initiated', size=14)
plt.title(label='Average friendship initiated by males and females', size=16)
```

# **Observations**
1)Male and female Users ratio is approximately 40:60.

2)Friend count for Female is higher than male average friends count.

3)Friendship initiated by Males are higher than Females.

# **3) Analysis based on Age Groups**

```
#Age_groups
#1)Maximum Number of users of particular age group
figure=plt.figure(figsize=[25,7])
data.age_group.value_counts()
Countuses_age_group = pd.value_counts(data['age_group'].values, sort=True)
print(Countuses_age_group)
Countuses_age_group.plot.pie(
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title(label='Percentage of Users of Different age groups', size=16)
```

```
#2)To analyse the friend_count in reference to age groups.
print(data.groupby(['age_group','gender'])['friend_count'].sum())
plt.rcParams["figure.figsize"] = (15,8)
data.groupby(['age_group','gender'])['friend_count'].sum().unstack(level=-1).plot(kind='bar',stacked=False, color=["green", "orange", "indianred"])
plt.legend(loc='upper left')
plt.ylabel(ylabel='Friends_count', size=14)
plt.title(label='Friends count of Different age groups', size=16)
plt.show()
```

```
#3)Which age group initiates friendship more
print(data.groupby(['age_group','gender'])['friendships_initiated'].sum())
plt.rcParams["figure.figsize"] = (10,5)
data.groupby(['age_group','gender'])['friendships_initiated'].sum().unstack(level=-1).plot(kind='bar',stacked=False, color=["darkcyan", "gold", "indianred"])
plt.legend(loc='upper left')
plt.ylabel(ylabel='Friendship_initiated', size=14)
plt.title(label='Friendships initiated by Different age groups', size=16)
plt.show()
```

```
#4) Most active age_group
import matplotlib.pyplot as plt
group_data1 = data.groupby(['age_group'])['friend_count'].sum()
group_data2 = data.groupby(['age_group'])['friendships_initiated'].sum()
group_data3 = data.groupby(['age_group'])['likes'].sum()
group_data4 = data.groupby(['age_group'])['likes_received'].sum()
plt.plot(group_data1)
plt.plot(group_data2)
plt.plot(group_data3)
plt.plot(group_data4)
plt.legend(["Friend_count", "friendships_initiated",'likes','likes_received'])
plt.ylabel("Frequency",size=14)
plt.xlabel("Age Groups",size=14)
plt.title(label='Activeness of Different Age group Users', size=16)
```

# **observations:**
1)Maximum users are of age group 21-30(28.9%) followed by age group 10-20(25.0%)

2)Friends count of age group 10-20 is highest followed by age group 21-30 .

3)Average Friendship initiated by age group 10-20 is highest followed by age group 21-30.

4) Most active age group is below 40.
